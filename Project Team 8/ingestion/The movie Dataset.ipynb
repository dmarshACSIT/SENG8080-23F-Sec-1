```python
import time
import requests

api_key = '48ce7f0a023cb3f3f4eb6027cff4f84f'

# Set the base URL for TMDB's API
base_url = 'https://api.themoviedb.org/3/'

# Fetch genre data
genre_data = {}
genre_endpoint = 'genre/movie/list'
genre_params = {'api_key': api_key}

start_time = time.time()
response = requests.get(base_url + genre_endpoint, params=genre_params)

if response.status_code == 200:
    genre_info = response.json()
    for genre in genre_info['genres']:
        genre_data[genre['id']] = genre['name']
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"INFO | Fetch Genre Transaction completed Successfully | Time taken: {elapsed_time} seconds")
else:
    print('ERROR | Error fetching genre data:', response.status_code)
```

```python
import requests


api_key = '48ce7f0a023cb3f3f4eb6027cff4f84f'


base_url = 'https://api.themoviedb.org/3/'


endpoint = 'discover/movie'
params = {
    'api_key': api_key,
    'primary_release_date.gte': '2020-01-01',
    'primary_release_date.lte': '2022-02-31',
}


all_movies = []
start_time = time.time()
page = 1
while True:
    params['page'] = page
    response = requests.get(base_url + endpoint, params=params)

    if response.status_code == 200:
        movie_data = response.json()
        if not movie_data['results']:
            break
        all_movies.extend(movie_data['results'])
        page += 1
    else:
        print('Error:', response.status_code)
        break

end_time = time.time()
elapsed_time = end_time - start_time
elapsed_time = round(elapsed_time, 3)
print(f"INFO | Fetch Movie Transaction completed Successfully | Time taken: {elapsed_time} seconds")

#print all the title of the movies
#for movie in all_movies:
#    print(movie['title'])

```

```python
#map the genre
genre_name_map = {genre_id: genre_name for genre_id, genre_name in genre_data.items()}

# Iterate through all movies and replace genre IDs with genre names
for movie in all_movies:
    genre_ids = movie.get('genre_ids', [])
    genre_names = [genre_name_map.get(genre_id, 'Unknown') for genre_id in genre_ids]
    movie['genres'] = genre_names
print("INFO | Mapping Genre IDs with Genre Name transaction completed successfully")
```

```python
#cleaning the data as multiple genres are getting populated in same column to deal with it we are splitting the genre in multiple columns
import pandas as pd
start_time = time.time()
# Assuming you've already retrieved 'all_movies' and mapped genre names to each movie

# Create a list of dictionaries, where each dictionary represents a movie
movies_data = []

# Create a set of unique genre names
unique_genres = set()

for movie in all_movies:
    title = movie.get('title', 'N/A')
    overview = movie.get('overview', 'N/A')
    release_date = movie.get('release_date', 'N/A')
    average_rating = movie.get('vote_average', 'N/A')
    runtime = movie.get('runtime', 'N/A')
    production_companies = ', '.join([company.get('name', 'N/A') for company in movie.get('production_companies', [])])
    languages = ', '.join([language.get('name', 'N/A') for language in movie.get('spoken_languages', [])])

    genre_ids = movie.get('genre_ids', [])
    genre_names = [genre_name_map.get(genre_id, 'Unknown') for genre_id in genre_ids]

    unique_genres.update(genre_names)

    movie_data = {
        'Title': title,
        'Overview': overview,
        'Release Date': release_date,
        'Average Rating': average_rating,
        'Runtime (minutes)': runtime,
        'Production Companies': production_companies,
        'Languages': languages,
    }

    for genre in genre_names:
        movie_data[genre] = 1  # Set a value of 1 for the genre column
    movies_data.append(movie_data)


# Create a DataFrame from the list of movie data
movie_df = pd.DataFrame(movies_data)

# Fill missing values with 0 (movies that don't belong to a particular genre)
movie_df.fillna(0, inplace=True)

# Reset the index of the DataFrame
movie_df.reset_index(drop=True, inplace=True)

end_time = time.time()
elapsed_time = end_time - start_time
elapsed_time = round(elapsed_time, 3)
print(f"INFO | Cleaning the Genre Transaction completed Successfully | Time taken: {elapsed_time} seconds")


```
