```python
#--------------------------Get The Movie list with specified Date range Code Starts Here ---------------------------
import requests
import time

api_key = '48ce7f0a023cb3f3f4eb6027cff4f84f'
base_url = 'https://api.themoviedb.org/3/'
endpoint = 'discover/movie'
params = {
    'api_key': api_key,
    'primary_release_date.gte': '2020-01-01',
    'primary_release_date.lte': '2020-01-31',
    'append_to_response': 'id'
}

all_movies = []
start_time = time.time()
page = 1
print("INFO | Fetching the list of movies from the TMDB API GET CALL")
print("INFO | Transaction Started")
while True:
    params['page'] = page
    response = requests.get(base_url + endpoint, params=params)

    if response.status_code == 200:
        movie_data = response.json()
        if not movie_data['results']:
            break
        all_movies.extend(movie_data['results'])
        page += 1
    else:
        print('Error:', response.status_code)
        break

end_time = time.time()
elapsed_time = end_time - start_time
elapsed_time = round(elapsed_time, 3)
print(f"INFO | Fetch Movie Transaction completed Successfully | Time taken: {elapsed_time} seconds")

# Print all the titles and IDs of the movies
#for movie in all_movies:
#    print(f"Title: {movie['title']}, ID: {movie['id']}")


#--------------------------Get The Movie list with specified Date range Code Ends Here ---------------------------
```

```python
#map the genre
genre_name_map = {genre_id: genre_name for genre_id, genre_name in genre_data.items()}

# Iterate through all movies and replace genre IDs with genre names
print("INFO | Mapping the Genres IDs with Genre Names")
for movie in all_movies:
    genre_ids = movie.get('genre_ids', [])
    genre_names = [genre_name_map.get(genre_id, 'Unknown') for genre_id in genre_ids]
    movie['genres'] = genre_names
print("INFO | Mapping Genre IDs with Genre Name transaction completed successfully")
```

```python
import pandas as pd
start_time = time.time()

# Assuming you've already retrieved 'all_movies' and mapped genre names to each movie

# Create a list of dictionaries, where each dictionary represents a movie
movies_data = []

# Create a set of unique genre names
unique_genres = set()

for movie in all_movies:
    movie_id = movie.get('id', 'N/A')  # Add movie ID
    title = movie.get('title', 'N/A')
    overview = movie.get('overview', 'N/A')
    release_date = movie.get('release_date', 'N/A')
    average_rating = movie.get('vote_average', 'N/A')
    runtime = movie.get('runtime', 'N/A')
    production_companies = ', '.join([company.get('name', 'N/A') for company in movie.get('production_companies', [])])
    languages = ', '.join([language.get('name', 'N/A') for language in movie.get('spoken_languages', [])])

    genre_ids = movie.get('genre_ids', [])
    genre_names = [genre_name_map.get(genre_id, 'Unknown') for genre_id in genre_ids]

    unique_genres.update(genre_names)

    movie_data = {
        'ID': movie_id,  # Include movie ID
        'Title': title,
        'Overview': overview,
        'Release Date': release_date,
        'Average Rating': average_rating,
        'Runtime (minutes)': runtime,
        'Production Companies': production_companies,
        'Languages': languages,
    }

    for genre in genre_names:
        movie_data[genre] = 1  # Set a value of 1 for the genre column
    movies_data.append(movie_data)

# Create a DataFrame from the list of movie data
movie_df = pd.DataFrame(movies_data)

# Fill missing values with 0 (movies that don't belong to a particular genre)
movie_df.fillna(0, inplace=True)

# Reset the index of the DataFrame
movie_df.reset_index(drop=True, inplace=True)

end_time = time.time()
elapsed_time = end_time - start_time
elapsed_time = round(elapsed_time, 3)
print(f"INFO | Cleaning, Genre Transaction completed Successfully | Time taken: {elapsed_time} seconds")
```

```python
import mysql.connector
import pandas as pd
import configparser

def insert_dataframe_to_mysql(df, table_name, conn):
    """Insert a pandas DataFrame into a MySQL table."""
    cursor = conn.cursor()

    # Making sure column names with spaces or reserved words are wrapped in backticks
    col_names = ['`' + str(col) + '`' for col in df.columns]

    for _, row in df.iterrows():
        sql_query = f"INSERT INTO {table_name} ({', '.join(col_names)}) VALUES ({', '.join(['%s' for _ in df.columns])})"
        try:
            cursor.execute(sql_query, tuple(row))
            print(f"Inserted data for movie {row['Title']} (ID: {row['ID']})")
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            print(f"Failed to insert data for movie {row['Title']} (ID: {row['ID']})")

    conn.commit()
    cursor.close()
    conn.close()

def init_conn(db_config):
    conn = mysql.connector.connect(**db_config)
    return conn

def load_config(config_file):
    config = configparser.ConfigParser()
    config.read(config_file)
    datasetPath = config['DatasetPaths']['movieDataset']
    db_config = {
        'host': config['dbconnection']['host'],
        'user': config['dbconnection']['user'],
        'password': config['dbconnection']['password'],
        'database': config['dbconnection']['database']
    }
    return db_config 
if __name__ == "__main__":
    try:
        config_file = 'appConfig.ini'
        db_config = load_config(config_file)
        conn = init_conn(db_config)
        insert_dataframe_to_mysql(movie_df,"movies",conn)
        connection.close()
         
    except Exception as e:
        print(f"An error occurred: {e}")
        print(traceback.print_exc())
```
